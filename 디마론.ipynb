## 네이버 사용의 목적 / 사용의 이유

```python
change= naver.loc[naver['네이버_사용목적'].str.contains('혜택')]
change['네이버_사용목적'].values
```

```python
# 구매 카테고리 빈도분석
naver_category=[]
for cat in naver['네이버_사용목적']:
  categories = cat.split(', ')
  for a in categories:
    naver_category.append(a)

category = pd.DataFrame(naver_category,columns=['사용목적'])
cat_gr= category.groupby('사용목적')[['사용목적']].size().reset_index(name='빈도수')

cat_gr.sort_values(by='빈도수', ascending=False, inplace=True)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(data=cat_gr,
              x='사용목적',
            y='빈도수',
            palette='deep')

plt.title('네이버의 사용 목적(중복허용)')
cat_gr


# 네이버 쇼핑 > 단순 쇼핑의 용도가 아님.
# 아 뭘 사야겠다 > 일단 네이버에 검색 > 당연히 네이버 쇼핑의 할인과 가격비교 정보 확인들이 부가적으로 들어옴
# 네이버라는 검색엔진의 이점이 다시 드러나는 것처럼 보인다.
# 사야돼! > 일단 네이버 검색 > 가격, 정보와같은 부가정보가 같이 들어옴 > 그 중에서 가장 혜택이 좋은 것을 구매 > 네이버 쇼핑을 사용하게됨.
# 위와 같은 순서로 구매가 이루어지지 않을까 생각이 들음
# 그래서 상위권에 가격 비교, 정보검색, 온라인의 편의성등이 올라오고 생각보다 쇼핑 목적의 구매는 빈도가 적은 것 같다고 볼 수 있을 수도.

# 그렇다면 쇼핑을 위해서는 무조건 네이버 쇼핑에가서 구매를 해야돼! 라는 생각이 들게하는게 좋은 포인트일 수도 있을 것 같음.
# 쇼핑 = 네이버 쇼핑에서 라는 공식을 성립시키는 방안을 보여줘야할듯.
```

### 그렇다면 사용을 많이하는 사람들의 실구매 횟수가 많은가?

```python
# 산점도 or 회귀 분석 위해 숫자로 변경해줌
# 예시 ) 이용 1~5회  = 1~5 사이의  중 랜덤으로 임의의 값을 부여해주었다.

import random
naver['네이버_1달사용횟수'] = naver['네이버_1달사용횟수'].apply(lambda x: random.randint(1, 5) if x == '1회~5회 이용' else
                                                                random.randint(6, 10) if x == '6~10회 이용' else
                                                                random.randint(11, 15) if x == '11~15회 이용' else
                                                                random.randint(15, 19))

# '네이버_1달월구매횟수' 컬럼 값 변경
naver['네이버_1달월구매횟수'] = naver['네이버_1달월구매횟수'].apply(lambda x: 0 if x == '0회 구매' else
                                                                    random.randint(1, 3) if x == '1~3회 구매' else
                                                                    random.randint(4, 6) if x == '4~6회 구매' else
                                                                    random.randint(7, 9) if x == '7~10회 구매' else
                                                                    random.randint(10, 12))
```

```python
sns.regplot(data=naver,
            x='네이버_1달사용횟수',
            y='네이버_1달월구매횟수')

plt.title('1달 사용횟수와 1달 구매횟수간의 분포도 비교')

# 어느정도의 선형관계가 보인다고 볼 수있다. 사실 많이 쓰면 많이 구매하게 되는 건 어느정도 상식수준에서 납득되는 점이라고 생각 가능.
# 데이터의 양이 부족하여 확실하다고 볼 수 없지만, 어느정도 선형관계라는 경향성은 파악, 예측할 수 있을 듯 하다.
# 랜덤값이기 때문에 대략 0.2 ~ 0.4 사이의 분포를 보임
```

```python
# 상관분석을 진행
corr= naver[['네이버_1달사용횟수','네이버_1달월구매횟수']].corr()
sns.heatmap(corr, annot=True,
            fmt='.2f')
# 0.38의 상관계수값. 약한 양의 상관관계가 존재한다.
# > 네이버의 경우 앱을 많이 쓰게 하는 걸 유도하는 방안 즉, 리텐션이나 재구매율을 높이려는 노력 필요
```



* 0.38의 상관계수값. 약한 양의 상관관계가 존재한다.
* 네이버의 경우 서비스를 많이 쓰게 하는 걸 유도하는 방안 즉, 리텐션이나 재구매율을 높이려는 노력 필요
