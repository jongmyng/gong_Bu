## 네이버 사용의 목적 / 사용의 이유

```python
change= naver.loc[naver['네이버_사용목적'].str.contains('혜택')]
change['네이버_사용목적'].values
```

```python
# 네이버 사용의 목적
naver_category=[]
for cat in naver['네이버_사용목적']:
  categories = cat.split(', ')
  for a in categories:
    naver_category.append(a)

category = pd.DataFrame(naver_category,columns=['사용목적'])
cat_gr= category.groupby('사용목적')[['사용목적']].size().reset_index(name='빈도수')

cat_gr.sort_values(by='빈도수', ascending=False, inplace=True)
```

### 테무의 전체만족도에 영향을 미치는 변수들의 순위(회귀분석)

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

train_X = both_copy_review_num[['둘다_네이버_신뢰도', '둘다_네이버_가격최저가',
                '둘다_네이버_후기/리뷰', '둘다_네이버_할인혜택','둘다_네이버_제품품질','둘다_네이버_배송']].to_numpy()

train_y= both_copy_review_num['둘다_네이버_전체만족도'].to_numpy()

lr.fit(train_X,train_y)
```

```python
coefficients = lr.coef_
features = both_copy_review_num[['둘다_네이버_신뢰도', '둘다_네이버_가격최저가',
                '둘다_네이버_후기/리뷰', '둘다_네이버_할인혜택','둘다_네이버_제품품질','둘다_네이버_배송']].columns


coef_df = pd.DataFrame({'변수': features, '계수값': coefficients})
coef_df['계수값'] = round(coef_df['계수값'],2)
coef_df.sort_values(by='계수값',ascending=False,inplace=True)

plt.figure(figsize=(8,5))
sns.barplot(data=coef_df,
            y='변수',
            x='계수값',
            palette='deep')

plt.title('둘다_네이버를 쓰는 사람들의 전체 만족도 영향을 미치는 변수들의 순위')
coef_df

# 배송이 빠른건 큰의미가 없다. 음의상관관계
# 배송이 빠르면 만족도가 떨어진다가 아니라
# 다른 신뢰도. 후기 리뷰값이 너무 중요해서 상대적으로 음수가 나온걸지도
```
